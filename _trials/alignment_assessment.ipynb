```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
results_ws = pd.read_csv("dtw_assessment_white_scale.csv")
results_spd = pd.read_csv("dtw_assessment_speed.csv")
results_spd
```

```python
import webbrowser
from pathlib import Path

def open_row(i, man=None, results=results_spd):
    filename = results.loc[i, "csv"].replace(".csv", "_{}.html".format("dtw" if man is None else man))
    path = str(Path("/home/tom/Documents/AutoJudge/fc_json_assessments") / filename )

    webbrowser.get('/usr/bin/google-chrome %s').open(path)

open_row(24, results=results_spd )
```

```python

def success_metrics(res):
    return dict(
        failure_percentage = 100 * len(res.loc[res.worst_elm_difference > 6]) / len(res),
        
    )


def filter_res( whiten, scaling): 
    return results_ws.loc[(results_ws.whiten==whiten) & (results_ws.scaling==scaling)]
    
print(success_metrics(filter_res(True, "measure")))
print(success_metrics(filter_res(True, "global")))
print(success_metrics(filter_res(False, "measure")))
print(success_metrics(filter_res(False, "global")))
```

```python
import plotly.express as px
px.scatter(results_spd, x="speed_factor", y="worst_elm_difference", color="schedule")
```

```python
sfs = results_spd.speed_factor.unique()
rspf = [dict(speed_factor=spf, **success_metrics(results_spd.loc[results_spd.speed_factor == spf])) for spf in sfs]    

dfspf=pd.DataFrame(data=rspf)

px.line(dfspf, x="speed_factor", y="failure_percentage", title="Template Speed Factor Against Alignment Process Failure Percentage")
```
