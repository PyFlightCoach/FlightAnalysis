```python
from flightanalysis import ManoeuvreAnalysis
from json import load

with open("examples/scoring/manoeuvre/ma.json", "r") as f:
    data = load(f)


ma = ManoeuvreAnalysis.from_dict(data)

from flightplotting import plotdtw

plotdtw(ma.aligned, ma.intended.elements).update_layout(width=600, height=600)

```

```python
res = ma.intended.analyse(ma.aligned, ma.intended_template)

eid=0
er = res[eid+1]


ea = ma[eid]

res.downgrade_df()
```

```python
from flightplotting.traces import vectors, axestrace, vector
from geometry import Quaternion


fig = ea.plot_3d(nmodels=3, scale=1)
fig.add_traces(axestrace(ea.coord, 5))
#fig.add_traces(vectors(5, ea.fl, 5*res.roll_rate.measurement.value.unit()*res.roll_rate.measurement.visibility ))

fig.update_layout(height=800, width=600)


fig.add_traces(vectors(50,ea.fl,er.track_y.measurement.value.degrees()))

fig.add_traces(vector(ea.fl.pos[-1],10*er.track_z.measurement.value.degrees()))
```
