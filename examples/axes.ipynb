```python
from flightanalysis import State, Box
from flightdata import Flight

flight = Flight.from_csv("examples/data/00000035.csv").flying_only()
box = Box.from_f3a_zone("examples/data/box.f3a")
bst = State.from_flight(flight, box)

```

```python
jst = bst.to_judging()

```

```python
from flightplotting.plots import plotsec
start, end = 206, 206.5
fig = plotsec(bst[start:end], scale=10, nmodels=1)
fig = plotsec(jst[start:end], scale=10, nmodels=1, fig=fig, color="red")
#fig = plotsec(wsec[start:end], scale=10, nmodels=20, fig=fig, color="blue")

fig.update_layout(height=600, width=1000)
```

```python
from flightanalysis import Line
from geometry import Transformation, P0, Euler, PX, PY, PZ, Point
import numpy as np
from flightanalysis.state.tools.conversions import convert_state

judging_axis = Line(100, 0).create_template(
    Transformation(P0(), Euler(np.pi, 0, 0)),
    30
)

stability_axis = convert_state(judging_axis, PZ(np.radians(20)))

body_axis = convert_state(stability_axis, PY(np.radians(30)))

```

```python
from flightplotting.plots import plotsec
start, end = 0,1
fig = plotsec(body_axis[start:end], scale=10, nmodels=1)
fig = plotsec(stability_axis[start:end], scale=10, nmodels=1, fig=fig, color="red")
fig = plotsec(judging_axis[start:end], scale=10, nmodels=1, fig=fig, color="blue")
#fig = plotsec(wsec[start:end], scale=10, nmodels=20, fig=fig, color="blue")

fig.update_layout(height=600, width=1000)
```

```python
from flightanalysis.model.flow import Flow
from flightanalysis.environment import Environment
from flightanalysis.environment.environment import Air

from flightanalysis.environment.wind import WindModel

env = Environment.from_constructs(
    body_axis.time, 
    atm=Air.iso_sea_level(len(body_axis)), 
    wind=P0(len(body_axis))
)

flow = Flow.build(body_axis, env)


np.degrees(flow[0].data.loc[:,["alpha", "beta"]])
```
